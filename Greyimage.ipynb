```python
kernel1 = np.array([[1/25,1/25,1/25],
                   [1/25,1/25,1/25],
                   [1/25,1/25,1/25]])
```

```python
blur=cv2.filter2D(src=image,ddepth=-1,kernel=kernel1)
blur_img=np.concatenate((blur,image),axis=1)
cv2_imshow(blur_img)
```

```python
kernel2=np.array([[-1,0,1],
                  [-2,0,2],
                  [-1,0,1]])
f_img=cv2.filter2D(src=image,ddepth=-1,kernel=kernel2)
cv2_imshow(f_img)
```
